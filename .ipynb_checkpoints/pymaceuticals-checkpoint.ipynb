```python
# Observations/inferences from the data
# (1) According to the box plot, mice under the Ceftamin and Infubinol regimen have a median final tumor volume of about 60mm3
# while mice under Capomulin and Ramicane have median final tumor volumes of 38.13mm3 and 36.56mm3 respectively.
# (2) According to the line plots, mice given Capomulin and Ramicane see a similar tumor volume decrease over time while those given
# Ceftamin and Infubinol see a consistent tumor volume increase. In other words, Ceftamin and Infubinol are drug regimens that appear 
# to have no to little effect on tumor volume reduction.
# (3) For the Capomulin regimen, mouse weight and average tumor volume have a correlation coefficient of 0.95, implying a strong
# positive correlation between the two.
```

```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
import numpy as np

# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

# Combine the data into a single dataset
mouse_merge = pd.merge(mouse_metadata, study_results, how = "outer", on = "Mouse ID")
# Display the data table for preview
mouse_merge
```

```python
# Checking the number of mice.
mouse_merge["Mouse ID"].nunique()
```

```python
# Getting the duplicate mice by ID number that shows up for Mouse ID and Timepoint. 
dupl_check = mouse_merge.duplicated(["Mouse ID", "Timepoint"], keep = False)
dupl_value = dupl_check.loc[dupl_check == True]

dupl_mouse_list = []
for n in range(len(dupl_value)):
    dupl_mouse_list.append(mouse_merge["Mouse ID"][dupl_value.index[n]])
dupl_mouse_list = list(set(dupl_mouse_list))
print(dupl_mouse_list)
```

```python
# Optional: Get all the data for the duplicate mouse ID. 
filtered_df = pd.DataFrame(columns=mouse_merge.columns)

dupl_values_list = []
for n in range(len(dupl_mouse_list)):
    dupl_values_list.append(mouse_merge.loc[mouse_merge["Mouse ID"] == dupl_mouse_list[n]])
filtered_df = filtered_df.append(dupl_values_list, ignore_index = True)
filtered_df
```

```python
# Create a clean DataFrame by dropping the duplicate mouse by its ID.
filtered_mouse = mouse_merge.copy()

for n in range(len(dupl_mouse_list)):
    filtered_mouse.drop(filtered_mouse[filtered_mouse["Mouse ID"] == dupl_mouse_list[n]].index, inplace=True)

filtered_mouse.reset_index(drop=True, inplace=True)
filtered_mouse
```

```python
# Checking the number of mice in the clean DataFrame.
filtered_mouse["Mouse ID"].nunique()
```

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen
# This method is the most straighforward, creating multiple series and putting them all together at the end.
unique_drugs = filtered_mouse["Drug Regimen"].unique() 
tumor_vol = filtered_mouse[["Drug Regimen", "Tumor Volume (mm3)"]]
summary_df = pd.DataFrame(columns = ["Drug Regimen", "Mean", "Median", "Variance", "Standard Deviation", "SEM"])

for n in unique_drugs:
    mean = tumor_vol.loc[tumor_vol["Drug Regimen"] == n].mean()
    median = tumor_vol.loc[tumor_vol["Drug Regimen"] == n].median()
    var = tumor_vol.loc[tumor_vol["Drug Regimen"] == n].var()
    sd = tumor_vol.loc[tumor_vol["Drug Regimen"] == n].std()
    sem = tumor_vol.loc[tumor_vol["Drug Regimen"] == n].sem()
    summary_df = summary_df.append({"Drug Regimen": n,
                                    "Mean": mean[0],
                                    'Median': median[0],
                                    "Variance": var[0],
                                    "Standard Deviation": sd[0],
                                    "SEM": sem[0]
                                   }, ignore_index=True)

summary_df = summary_df.set_index("Drug Regimen")
summary_df
```

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen
# This method produces everything in a single groupby function
summary_df = filtered_mouse.groupby("Drug Regimen", sort=False)

group_mean = summary_df["Tumor Volume (mm3)"].mean()
group_median = summary_df["Tumor Volume (mm3)"].median()
group_var = summary_df["Tumor Volume (mm3)"].var()
group_sd = summary_df["Tumor Volume (mm3)"].std()
group_sem = summary_df["Tumor Volume (mm3)"].sem()

summary_df_groupby = pd.DataFrame(group_mean)
summary_df_groupby = summary_df_groupby.rename(columns = {"Tumor Volume (mm3)":"Mean"})
summary_df_groupby["Median"] = group_median
summary_df_groupby["Variance"] = group_var
summary_df_groupby["Standard Deviation"] = group_sd
summary_df_groupby["SEM"] = group_sem

summary_df_groupby
```

```python
# Generate a bar plot showing the total number of mice for each treatment throughout the course of the study using pandas. 
summary_df_groupby = pd.DataFrame(summary_df["Mouse ID"].count())

mice_plot_pandas = summary_df_groupby.plot(kind = "bar", title = "Number of Mice by Drug Regimen", figsize=(20,5), legend = False, 
                                           ylim = (0, max(summary_df_groupby["Mouse ID"])+25))
mice_plot_pandas.set_xticklabels(summary_df_groupby.index, rotation = 45)
mice_plot_pandas.set_ylabel("Number of Mice")

plt.savefig("plot_images/number_mice.png")
plt.show()
```

```python
# Generate a bar plot showing the total number of mice for each treatment throughout the course of the study using pyplot.
x_axis = np.arange(len(summary_df_groupby))
tick_locations = [value for value in x_axis]

plt.figure(figsize=(20, 5))
plt.bar(x_axis, summary_df_groupby["Mouse ID"], width = 0.5)
plt.xticks(tick_locations, summary_df_groupby.index, rotation = 45)
plt.title("Number of Mice by Drug Regimen")
plt.xlabel("Drug Regimen")
plt.ylabel("Number of Mice")
plt.xlim(-0.5, len(x_axis)-0.5)
plt.ylim(0, max(summary_df_groupby["Mouse ID"])+25)

plt.show()
```

```python
# Calculate the final tumor volume of each mouse across four of the treatment regimens:  
# Capomulin, Ramicane, Infubinol, and Ceftamin

# Start by getting the last (greatest) timepoint for each mouse
# Merge this group df with the original dataframe to get the tumor volume at the last timepoint

# Capomulin
capo_data = filtered_mouse.loc[(filtered_mouse["Drug Regimen"] == "Capomulin")]
# find the max timepoint of each mouse
capo_max_time = capo_data.groupby("Mouse ID")["Timepoint"].max()
# convert to dataframe
capo_max_time = capo_max_time.reset_index()
# merge with the original dataframe
capo_final_tumor = pd.merge(capo_max_time, filtered_mouse, on = ("Mouse ID", "Timepoint"), how = "left")

print("Capomulin:")
print(capo_final_tumor.head())
print()

# Ceftamin
ceft_data = filtered_mouse.loc[(filtered_mouse["Drug Regimen"] == "Ceftamin")]
ceft_max_time = ceft_data.groupby("Mouse ID")["Timepoint"].max()
ceft_max_time = ceft_max_time.reset_index()
ceft_final_tumor = pd.merge(ceft_max_time, filtered_mouse, on = ("Mouse ID", "Timepoint"), how = "left")

print("Ceftamin:")
print(ceft_final_tumor.head())
print()

# Infubinol
infu_data = filtered_mouse.loc[(filtered_mouse["Drug Regimen"] == "Infubinol")]
infu_max_time = infu_data.groupby("Mouse ID")["Timepoint"].max()
infu_max_time = infu_max_time.reset_index()
infu_final_tumor = pd.merge(infu_max_time, filtered_mouse, on = ("Mouse ID", "Timepoint"), how = "left")

print("Infubinol:")
print(infu_final_tumor.head())
print()

# Ramicane
rami_data = filtered_mouse.loc[(filtered_mouse["Drug Regimen"] == "Ramicane")]
rami_max_time = rami_data.groupby("Mouse ID")["Timepoint"].max()
rami_max_time = rami_max_time.reset_index()
rami_final_tumor = pd.merge(rami_max_time, filtered_mouse, on = ("Mouse ID", "Timepoint"), how = "left")

print("Ramicane:")
print(rami_final_tumor.head())
```

```python
# Put treatments into a list for for loop (and later for plot labels)
treatments = ["Capomulin", "Ceftamin", "Infubinol", "Ramicane"]
# Create empty list to fill with tumor vol data (for plotting)
tumor_vol_plot = []

# Calculate the IQR and quantitatively determine if there are any potential outliers. 
    # add subset 

# Capomulin
# Get the final tumor volumes
capo_vol = capo_final_tumor['Tumor Volume (mm3)']
# Add to tumor_vol_plot list
tumor_vol_plot.append(capo_vol)
# Calculate upper and lower bounds
quartiles = capo_vol.quantile([.25,.5,.75])
lowerq = quartiles[0.25]
upperq = quartiles[0.75]
iqr = upperq-lowerq
print(f"The interquartile range of Capomulin tumor volume data is: {round(iqr, 2)}.")
lower_bound = lowerq - (1.5*iqr)
upper_bound = upperq + (1.5*iqr)
print(f"Capomulin: values below {round(lower_bound, 2)} and above {round(upper_bound, 2)} could be outliers.")

# Find outliers
capo_outlier = capo_final_tumor.loc[(capo_vol < lower_bound) | (capo_vol > upper_bound)]
if capo_outlier.empty == True:
    print("There are no outliers.")
else:
    print("Outlier(s):")
    for n in range(len(capo_outlier)):
        print("Mouse ID " + str(capo_outlier["Mouse ID"][n]) + " with a tumor volume of " + 
              str(round(capo_outlier["Tumor Volume (mm3)"][n], 2)) +".")
```

```python
# Repeat the previous step for other drug regimens
# Ceftamin
ceft_vol = ceft_final_tumor['Tumor Volume (mm3)']
tumor_vol_plot.append(ceft_vol)

quartiles = ceft_vol.quantile([.25,.5,.75])
lowerq = quartiles[0.25]
upperq = quartiles[0.75]
iqr = upperq-lowerq
print(f"The interquartile range of Ceftamin tumor volume data is: {round(iqr, 2)}.")
lower_bound = lowerq - (1.5*iqr)
upper_bound = upperq + (1.5*iqr)
print(f"Ceftamin: values below {round(lower_bound, 2)} and above {round(upper_bound, 2)} could be outliers.")

# outliers
ceft_outlier = ceft_final_tumor.loc[(ceft_vol < lower_bound) | (ceft_vol > upper_bound)]
if ceft_outlier.empty == True:
    print("There are no outliers.")
else:
    print("Outlier(s):")
    for n in range(len(ceft_outlier)):
        print("Mouse ID " + str(ceft_outlier["Mouse ID"][n]) + " with a tumor volume of " + 
              str(round(ceft_outlier["Tumor Volume (mm3)"][n], 2)) +".")
```

```python
# Infubinol
infu_vol = infu_final_tumor["Tumor Volume (mm3)"]
tumor_vol_plot.append(infu_vol)

quartiles = infu_vol.quantile([.25,.5,.75])
lowerq = quartiles[0.25]
upperq = quartiles[0.75]
iqr = upperq-lowerq
print(f"The interquartile range of Infubinol tumor volume data is: {round(iqr, 2)}.")
lower_bound = lowerq - (1.5*iqr)
upper_bound = upperq + (1.5*iqr)
print(f"Infubinol: values below {round(lower_bound, 2)} and above {round(upper_bound, 2)} could be outliers.")

# outliers
infu_outlier = infu_final_tumor.loc[(infu_vol < lower_bound) | (infu_vol > upper_bound)]
infu_outlier = infu_outlier.reset_index(drop = True)

if infu_outlier.empty == True:
    print("There are no outliers.")
else:
    print("Outlier(s):")
    for n in range(len(infu_outlier)):
        print("Mouse " + str(infu_outlier["Mouse ID"][n]) + " with a tumor volume of " + 
              str(round(infu_outlier["Tumor Volume (mm3)"][n], 2)) +".")
```

```python
# Ramicane
rami_vol = rami_final_tumor['Tumor Volume (mm3)']
tumor_vol_plot.append(rami_vol)

quartiles = rami_vol.quantile([.25,.5,.75])
lowerq = quartiles[0.25]
upperq = quartiles[0.75]
iqr = upperq-lowerq
print(f"The interquartile range of Ramicane tumor volume data is: {round(iqr, 2)}.")
lower_bound = lowerq - (1.5*iqr)
upper_bound = upperq + (1.5*iqr)
print(f"Ramicane: values below {round(lower_bound, 2)} and above {round(upper_bound, 2)} could be outliers.")

# outliers
rami_outlier = rami_final_tumor.loc[(rami_vol < lower_bound) | (rami_vol > upper_bound)]
if rami_outlier.empty == True:
    print("There are no outliers.")
else:
    print("Outlier(s):")
    for n in range(len(rami_outlier)):
        print("Mouse ID " + str(rami_outlier["Mouse ID"][n]) + " with a tumor volume of " + 
              str(round(rami_outlier["Tumor Volume (mm3)"][n], 2)) +".")
```

```python
# Generate a scatter plot of mouse weight versus average tumor volume for the Capomulin regimen
capo_data = capo_data.groupby(["Drug Regimen", "Weight (g)"])["Tumor Volume (mm3)"].mean()
capo_data = capo_data.reset_index()
x_scatter = capo_data["Weight (g)"]
y_scatter = capo_data["Tumor Volume (mm3)"]
plt.scatter(x_scatter,y_scatter, color = 'brown')
for x, y in zip(x_scatter, y_scatter):
    label = "{:.2f}".format(y)
    plt.annotate(label,
                (x,y),
                 textcoords = "offset points",
                 xytext = (0,5),
                 ha = "center")
plt.title("Average Tumor Volume vs. Mouse Weight for Capomulin Regimen")
plt.xlabel("Mouse Weight")
plt.ylabel("Tumor Volume (mm3)")
plt.grid(axis = "y")
plt.ylim(capo_tumor_vol_line["Tumor Volume (mm3)"].min()-0.5, capo_tumor_vol_line["Tumor Volume (mm3)"].max()+0.6)
plt.savefig("plot_images/tumor_vol_weight_scatter.png")
plt.show()
```

```python
# Calculate the correlation coefficient and linear regression model 
# for mouse weight and average tumor volume for the Capomulin regimen
correlation = st.pearsonr(capo_data["Weight (g)"], capo_data["Tumor Volume (mm3)"])
print(f"The correlation between mouse weight and average tumor volume for the Capomulin regimen is {round(correlation[0], 2)}.")

(capo_s, capo_int, capo_r, capo_p, capo_stderr) = st.linregress(capo_data["Weight (g)"], capo_data["Tumor Volume (mm3)"])
regress_values = capo_data["Weight (g)"] * capo_s + capo_int
line_eq = "y = " + str(round(capo_s, 2)) + "x + " + str(round(capo_int, 2))
print(f"The linear regression line is {line_eq}.")

plt.scatter(x_scatter, y_scatter, color = 'brown')
plt.plot(capo_data["Weight (g)"], regress_values, "g-")
plt.annotate(line_eq,(21, 41), fontsize = 12, color = "green")
plt.title("Average Tumor Volume vs. Mouse Weight for Capomulin Regimen")
plt.xlabel("Mouse Weight")
plt.ylabel("Tumor Volume (mm3)")
plt.grid(axis = "y")
plt.savefig("plot_images/tumor_vol_weight_linreg.png")
plt.show()
```
